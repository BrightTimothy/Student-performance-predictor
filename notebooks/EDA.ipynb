UNDERSTANDING MY TARGET VARIABLE: FINAL_RESULT OF THE STUDENT_INFO DATASET

```python
student_info["final_result"].unique()
```

```python
student_info["final_result"].value_counts()
```

In my target variable, there is class imabalance and training my model on this imbalance will make it lazy and always predict from the majority class most of the time. 
I have four distinct final_result classes and the goal of my student performance predictor is to provide an early warning system for advisors and not to predict if a student passed or failed..
So i will binarize my target variable, grouping the four distinct classes into two(at risk or safe)
Pass and distinction = 0 (safe)
Fail and withdrawn = 1(at risk)

VISUALIZING MY TARGET VARIABLE

```python
plt.figure()
ax = student_info["final_result"].value_counts().plot(kind= "bar", color = ["green", "red", "red", "green"])
for p in ax.patches:
    ax.annotate(str(p.get_height()), 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', 
                va='bottom')
plt.title("Count of Students Against their Final Result")
plt.xlabel("Final Result")
plt.ylabel("Count")
plt.xticks(rotation = 0)
plt.show()
```

From my visualization:
Total students that are safe = 12361 + 3024 = 15,385
Total students at risk = 17,208 
The divide seems quite balanced to me, so I move on. It is a 52.8% to 47.2%

EXPLORING THE INDIVIDUAL COLUMNS ACROSS MY STUDENT_INFO TABLE

```python
student_info["num_of_prev_attempts"].value_counts()
```

That means roughly 87% of students are first-timers and only 13% are retaking. So does it mean that students that retook the class have a higher chance of being safe or being at the at risk group

Cross referencing num_of_prev_attempts with final_result column of the Student_info dataset

```python
pd.crosstab(student_info["num_of_prev_attempts"], student_info["final_result"])
```

```python
pd.crosstab(student_info["num_of_prev_attempts"], student_info["final_result"], normalize= "index").mul(100).round(1)
```

"num_of_prev_attempts is a strong predictor of student outcome. As attempts increase, Fail and Withdrawn rates rise significantly while Distinction rates drop to zero. This column will be a key feature in the model."

```python
fig, ax = plt.subplots()
pd.crosstab(student_info["num_of_prev_attempts"], student_info["final_result"], normalize= "index").mul(100).round(1).plot(kind = "bar", ax=ax, color = ["darkgreen", "darkred", "limegreen", "red"])
plt.xlabel("Number of Previous Attempts")
plt.ylabel("Percentage %")
plt.title("Number of Attempts Vs Final Result")
plt.xticks(rotation = 0)
plt.show()
```

From the visualizations, it shows that as the number of attempts increases, the failed and withdrew rate increased.

ANALYSING AGE_BAND COLUMN OF STUDENT_INFO DATASET

```python
student_info["age_band"].value_counts()
```

```python
pd.crosstab(student_info["age_band"], student_info["final_result"])
```

```python
pd.crosstab(student_info["age_band"], student_info["final_result"], normalize = "index").mul(100).round(1)
```

age_band shows a clear relationship with outcomes. Older students (55+) have nearly double the Distinction rate of younger students (0-35) and significantly lower Fail and Withdrawn rates. This is likely driven by motivation and self-selection rather than cognitive ability. age_band will be a useful feature for the model.

```python
fig, ax = plt.subplots()
pd.crosstab(student_info["age_band"], student_info["final_result"], normalize= "index").mul(100).round(1).plot(kind = "bar", ax = ax, color = ["darkgreen", "darkred", "limegreen", "red"])
plt.xlabel("Age Bands")
plt.ylabel("Percentage of Students")
plt.title("Showing Age bands across Students")
plt.xticks(rotation = 0)
plt.show()
```
