In my target variable, there is class imabalance and training my model on this imbalance will make it lazy and always predict from the majority class most of the time. 
I have four distinct final_result classes and the goal of my student performance predictor is to provide an early warning system for advisors and not to predict if a student passed or failed..
So i will binarize my target variable, grouping the four distinct classes into two(at risk or safe)
Pass and distinction = 0 (safe)
Fail and withdrawn = 1(at risk)

The most important on this dataset is sum_click column because it shows how many times a particular student clicked on a particular resource on a platform and this shows total engagement. This is useful for a early warning signal. 

```python
pd.set_option("display.float_format", "{:.2f}".format)
student_Vle["sum_click"].describe()
```

```python
plt.hist(student_Vle["sum_click"], bins = 50, range=(1, 20))
plt.xlabel("Number of Clicks")
plt.ylabel("Frequency")
plt.title("Distribution of Clicks Per Activity")
plt.xticks(rotation = 0)
plt.show()
```

Most students make very few clicks on any single resource on any given day. A small number of students are highly engaged and click many times. Most students make very few clicks on any single resource on any given day. A small number of students are highly engaged and click many times.

VLE clicks must be aggregated at student + module + presentation level, not just student level, to avoid mixing clicks across different modules.

```python
total_clicks = student_Vle.groupby(["id_student", "code_module", "code_presentation"])["sum_click"].sum()
total_clicks
```

```python
student_Vle
```

```python
total_clicks.describe()
```

Look at the gap between the bottom 25% and the top 25%:

Bottom quarter → 261 clicks or fewer
Top quarter → 1,770 clicks or more

That's nearly a 7x difference in engagement between low and high engagement students.
And remember — low engagement strongly correlates with withdrawal and failure. A student making only 261 clicks across an entire semester is barely showing up digitally.

```python
student_info
```

Now I will need to merge the total_clicks table with the student_info table to see if my assumption is correct. But before that I will need to reset the index of my total_clicks column so pandas can identify them as columns rather than indexes.

```python
total_clicks = total_clicks.reset_index()
total_clicks
```

```python
student_info.columns
```

```python
clicks_vs_results = total_clicks.merge(
    student_info[["id_student", "code_module", "code_presentation", "final_result"]],
    on= ["id_student", "code_module", "code_presentation"],
    how= "left"
)

clicks_vs_results.head()
```

Even in 5 rows the pattern is already visible — lower clicks correlate with worse outcomes.

```python
average_clicks = clicks_vs_results.groupby("final_result")["sum_click"].mean().round(0).sort_values(ascending= False)
average_clicks
```

Total VLE clicks per student per module is a strong predictor of outcome. Distinction students average 2,667 clicks vs only 445 for Withdrawn students — a 6x difference. This will be a key feature in the model.

```python
fig, ax = plt.subplots()
average_clicks.plot(kind = "bar", ax = ax, color = ["darkgreen", "limegreen", "darkred", "red"])
plt.xlabel("Final Result")
plt.ylabel("Average Clicks")
plt.title("Total Clicks Against Final Result")
plt.xticks(rotation= 0)
plt.show()
```

From the chart, we can now see that the total clicks per student, per presentation, per module affects the overall result of the student, this is an important feature in building my model
